```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
customers = {
    "Customer_A": [50000, 25, 4.5, 10],
    "Customer_B": [52000, 27, 4.6, 11],
    "Customer_C": [30000, 15, 4.0, 7],
    "Customer_D": [80000, 40, 4.8, 15]
}
```

```python
def euclidean_distance(x, y):
    return np.sqrt(np.sum((np.array(x) - np.array(y)) ** 2))
```

```python
def manhattan_distance(x, y):
    return np.sum(np.abs(np.array(x) - np.array(y)))
```

```python
def cosine_sim(x, y):
    return cosine_similarity([x], [y])[0][0]
```

```python
names = list(customers.keys())

for i in range(len(names)):
    for j in range(i + 1, len(names)):

        c1 = customers[names[i]]
        c2 = customers[names[j]]

        euc = euclidean_distance(c1, c2)
        man = manhattan_distance(c1, c2)
        cos = cosine_sim(c1, c2)

        print("===================================")
        print(names[i], "vs", names[j])
        print("Euclidean Distance :", round(euc, 2))
        print("Manhattan Distance :", round(man, 2))
        print("Cosine Similarity  :", round(cos, 4))
```

```python
purchase_patterns = {
    "Customer_A": "ELECTRONICSBOOKS",
    "Customer_B": "ELECTRONICSCLOTHES",
    "Customer_C": "BOOKSGROCERY",
    "Customer_D": "ELECTRONICSBOOKS"
}
```

```python
def edit_distance(str1, str2):

    m = len(str1)
    n = len(str2)

    dp = [[0 for j in range(n + 1)] for i in range(m + 1)]

    for i in range(m + 1):
        dp[i][0] = i

    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):

            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]

            else:
                dp[i][j] = 1 + min(
                    dp[i - 1][j],
                    dp[i][j - 1],
                    dp[i - 1][j - 1]
                )

    return dp[m][n]
```

```python
names = list(purchase_patterns.keys())

print("\nEDIT DISTANCE RESULTS\n")

for i in range(len(names)):
    for j in range(i + 1, len(names)):

        d = edit_distance(
            purchase_patterns[names[i]],
            purchase_patterns[names[j]]
        )

        print(names[i], "vs", names[j], "=", d)
```
